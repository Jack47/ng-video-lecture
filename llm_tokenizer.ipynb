```python
text = "Ｕｎｉｃｏｄｅ! 🅤🅝🅘🅒🅞🅓🅔‽ 🇺‌🇳‌🇮‌🇨‌🇴‌🇩‌🇪! 😄 The very name strikes fear and awe into the hearts of programmers worldwide. We all know we ought to “support Unicode” in our software (whatever that means—like using wchar_t for all the strings, right?). But Unicode can be abstruse, and diving into the thousand-page Unicode Standard plus its dozens of supplementary annexes, reports, and notes can be more than a little intimidating. I don’t blame programmers for still finding the whole thing mysterious, even 30 years after Unicode’s inception."
bytes_tokens = text.encode('utf-8') # raw bytes: '\xef\xbc'
# byte to int
tokens = list(map(lambda x: int(x), bytes_tokens))
print('---')
print(f"{text=}\n{len(text)=}")
print('---')
print(f"{bytes_tokens=}\n{len(bytes_tokens)=}")
print('---')
print(f"{tokens=}\n{len(tokens)=}")
```

```python
# 遍历 ids 里相邻的两个元素，统计两两元素出现的次数
def get_stats(ids):
    counts = {}
    # pair 是一个元组，比如 (102, 32)
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1 
    return counts

stats = get_stats(tokens)
print(f"{stats=}")
print(f"{sorted(stats.items(), key=lambda x:x[1], reverse=True)=}")
```

```python
top_pair = max(stats, key=lambda x:x[1])
print(f"{top_pair=}")
top_pair = max(stats, key=stats.get)
print(f"{top_pair=}")
# 把  ids 里 pair 这个连续的两个元素，合并为一个（用新的 idx）
def merge(ids, pair, idx):
    new_ids = []
    i = 0
    while i <= len(ids)-1:
        if i < len(ids)-1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            new_ids.append(idx)
            i += 2
        else:
            new_ids.append(ids[i])
            i += 1
    
    return new_ids

print(f"{merge(ids=[5,6,6,7,9,1], pair=(6,7), idx=99)=}")

print(f"{len(merge(tokens, top_pair, 256))=}")
print(f"{merge(tokens, top_pair, 256)=}")
```

```python
# ---
vocab_size = 276 
num_merges = vocab_size - 256 # unicode 里最大的
ids = list(tokens) # copy
merges = {} # (int, int) -> int
for i in range(num_merges):
    idx = 256+i
    stats = get_stats(ids)
    top_pair = max(stats, key=stats.get)
    print(f"merging {top_pair} into a new token {idx}")
    ids = merge(ids=ids, pair=top_pair, idx=idx)
    merges[top_pair] = idx
```
